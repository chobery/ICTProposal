```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
text = "This is a sample sentence"
encoded_input = tokenizer.encode(text, return_tensors='pt')
import random

text = "This is a sample sentence"
encoded_input = tokenizer.encode(text, return_tensors='pt')

output = model.generate(encoded_input, max_length=600, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)

# 複数の文章をデコードして、ランダムに1つを選ぶ
decoded_outputs = tokenizer.batch_decode(output, skip_special_tokens=True)
random_output = random.sample(decoded_outputs, 1)[0]

print(random_output)
```
